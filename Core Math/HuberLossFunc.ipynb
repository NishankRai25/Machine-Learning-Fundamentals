```python
# This cost function is best of both worlds use when data has medium variance

def hubar_loss_function(x, y, w, b, a):
    n = len(x)
    cost_func = 0

    for i in range(n):
        f = w * x[i] + b
        if abs(f-y[i]) <= a:
            cost = (f-y[i]) ** 2
        else:
            cost = abs(f-y[i]) 
        cost_func += cost

    total_cost = 1/(2*n) * cost_func
    return total_cost
```
