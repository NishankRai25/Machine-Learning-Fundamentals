```python
# This cost fucntion is harsh to outliers, use when data has less variance

def mse_function(x, y, w, b):
    n = len(x)
    cost_func = 0

    for i in range(n):
        f = w * x[i] + b
        cost = (f-(y[i])) ** 2
        cost_func += cost

    total_cost = 1/(2*n) * cost_func
    return total_cost
```
