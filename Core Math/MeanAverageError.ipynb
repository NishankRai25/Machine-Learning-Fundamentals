```python
# This cost function is not harsh to outliers, use when data has a lots of variance

def mae_function(x, y, w, b):
    n = len(x)
    cost_func = 0

    for i in range(n):
        f = w * x[i] + b
        cost = abs(f-y[i])
        cost_func += cost

    total_cost = 1/(2*n) * cost_func
    return total_cost
```
