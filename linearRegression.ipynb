# Linear Regression
Linear Regression is a Supervised Machine Learning Algorithm

You feed the Computer, input x and output y, and with the data the computer makes a regression line that allows you to predict any out put y with an input x.

This is useful for housing prices, stock pricing and a lot more

This uses a cost function, the cost function is used to determine error in line, you want a regression line that is best fit for all the data points, so you most minimize error, When cost function is plotted, the graph is a parabula for 1 variable linear regression, a parabula has a minma. Gradient descent takes derivative of cost function, and subrtracts it from a learning rate, so we slowly approach the minma. This is how we get a best fit line


```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Housing.csv') #Loads Housing data
price = df["price"].values #Creates a vector
area = df[["area"]].values


df = df.drop(columns = ['bedrooms', 'bathrooms', 'stories', 'mainroad',
       'guestroom', 'basement', 'hotwaterheating', 'airconditioning',
       'parking', 'prefarea', 'furnishingstatus' #Keeps it only x and y

       ])
```

```python
# Cost Loss Function (Not used but useful to know)

def loss_function(m, b, points):
    total_error = 0 
    for i in range(len(points)): #the for loop is basically the Sum function
        x = points.iloc[i].area #Finds the actual x value
        y = points.iloc[i].price #Finds the actual y value
        
        total_error += (y - (m*x+b)) ** 2 #determines error by subtracting real value by predicted and squaring
    total_error / float(len(points))
```

```python
# Graident descent (minaimizing Loss Function)
# Gradient descent is used to minamize the loss fucntion, it does this by finding a Minma in the parbula and using an partial derivatives and learning fucntion to slowly approach minma


def graident_descent(m_now, b_now, points, l):
    m_gradient = 0 # Graident Descent for m
    b_gradient = 0 # Gradient descent for b

    n = len(points)
    
    for i in range(n): # Sum function 
        x = points.iloc[i].area # Real x
        y = points.iloc[i].price # Real y

        m_gradient += -(2/n) * x * (y-(m_now * x + b_now)) # Redoes cost function now, take partial derivative, and scales it
        b_gradient += -(2/n) * (y-(m_now * x + b_now))

    m = m_now - m_gradient * l #multiples by learning curve
    b = b_now - b_gradient * l

    return m, b
```

```python
m = 0
b = 0
l = 0.0000000001
epochs = 10000

for i in range(epochs): #Does it 10000 times do its perfect
    if i % 50 == 0:
        print(f"Epoch: {i}")
    m, b = graident_descent(m, b, df, l)


print(m, b)

plt.scatter(area, price, color="black")
plt.plot(list(range(0, 16200)), [m*x+b for x in range(0, 16200)], color ="red")
plt.show()
```
